```python
#频繁项集：满足最小支持度计数的项集
#强规则：同时满足最小支持度计数和最小置信度的项集，即频繁项集中满足最小置信度的项集
#关联规则挖掘：第一步找出所有频繁项集，第二步从频繁项集中找出强规则
#第一步的计算量巨大，为了克服这一困难，引入闭频繁项集和极大频繁项集的概念
#闭项集：对项集X而言，添加一项就会使该项集支持度计数减少，则X为闭频繁项集
#闭频繁项集：即项集X既是闭的又是频繁的
#极大频繁项集：对频繁项集X而言，添加一项就会使该项集不频繁，则X为极大频繁项集
```

```python
%%latex
$Apriori算法\\
Apriori算法采用逐层搜索的方法，频发k项集用于找出频繁k+1项集，为了提高\\
搜索效率，运用先验性质：频繁项集的非空子集也是频繁的。\\
Apriori步骤:\\
一、从D中产生1项集集合C_1\\
二、从C_1中产生频繁1项集集合L_1\\
三、连接步：L_1自连接，产生2项集集合C_2\\
四、剪枝步：从C_2中产生频繁2项集集合L_2\\
五、重复三、四步骤，直到产生频繁k项集集合L_k\\
六、从频繁k项集中产生强规则\\
连接步：自连接，且要保证前k-2项相同。\\
例如：要从L_2\{\{11,12\},\{11,13\},\{12,14\}\}中产生C_3，L_2自连接，要保证前1项\\
相同，所以C_3=\{\{11,12\}U\{11,13\}\}=\{11,12,13\}\\
注意：没有涉及到14，因为要保证前1项相同，14前1项是12，L_2自连接不能产生包含14的3项集\\
产生强规则：若最终产生了频繁4项集L_4{1，2，3，4}\\
则L_4的子集为\{\{1\},\{2\},\{3\},\{4\},\{1,2,3\},\{1,2,4\},\{1,3,4\},\{2,3,4\}\}\\
注：只需要1项集和k-1项集，因为如果想要其他项集如2项集的关联规则，只需要生成频繁3项集就可以了。\\
从L_4中产生所有规则：\\
\{\{\{1\}\Rightarrow\{2,3,4\}\},\\
  \{\{2\}\Rightarrow\{1,3,4\}\},\\
  \{\{3\}\Rightarrow\{1,2,4\}\},\\
  \{\{4\}\Rightarrow\{1,2,3\}\},\\
  \{\{1,2,3\}\Rightarrow\{4\}\},\\
  \{\{1,2,4\}\Rightarrow\{3\}\},\\
  \{\{1,3,4\}\Rightarrow\{2\}\},\\
  \{\{2,3,4\}\Rightarrow\{1\}\}\}\\
从中选出满足置信度的强规则。$
```
