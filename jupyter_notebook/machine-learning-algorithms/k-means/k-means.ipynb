```python
import numpy as np
from sklearn.datasets import make_blobs
n_samples = 1500
random_state = 170
X,y = make_blobs(n_samples,random_state = random_state)
```

```python
from sklearn.cluster import KMeans
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(12,12))
estimator = KMeans(n_clusters=2,random_state=random_state)
estimator.fit(X)
y_predicted = estimator.predict(X)
plt.subplot(221)
plt.scatter(X[:,0],X[:,1],c = y_predicted)
plt.title('Incorrect number of clusters')
transformation = [[ 0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X,transformation)
y_predicted_aniso = KMeans(n_clusters=3,random_state=random_state).fit_predict(X_aniso)
plt.subplot(222)
plt.scatter(X_aniso[:,0],X_aniso[:,1],c = y_predicted_aniso)
plt.title('Anisotropicly distributed data')
plt.subplot(223)
X_varied,y_varied = make_blobs(n_samples,cluster_std=[1.0,2.5,0.5],random_state=random_state)
y_predicted_varied = KMeans(n_clusters=3,random_state=random_state).fit_predict(X_varied)
plt.scatter(X_varied[:,0],X_varied[:,1],c = y_predicted_varied)
plt.title('Data with different std')
plt.subplot(224)
X_uneven = np.vstack((X[y==0][:500],X[y==1][:100],X[y==2][:10]))
y_predicted_uneven = KMeans(n_clusters=3,random_state=random_state).fit_predict(X_uneven)
plt.scatter(X_uneven[:,0],X_uneven[:,1],c = y_predicted_uneven)
plt.title('Uneven data')
plt.show()
```
