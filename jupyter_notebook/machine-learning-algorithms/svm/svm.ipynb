```python
#scikit-learn中svm可用于分类和回归
#svm分类器有三种：SVC,NuSVC,LinearSVC
#SVC,NuSVC类似，都是非线性分类器，可通过参数kernel指定非线性映射核函数
#'linear'(线性核函数), 'poly', 'rbf'(高斯径向基函数核函数), 'sigmoid', 'precomputed' 
#LinearSVC是线性分类器，没有kernel参数，默认用线性核函数
#只有用线性核函数的分类器才有coef_属性
#SVC可通过参数class_weight指定各类权重解决不平衡数据分类问题
#对svm分类器,可用sample_weight参数指定样本权重
#clf.decicion_function类似clf.predict,但返回的是待测试数据与分离超平面的距离而不是数据类别
#可以通过clf.decision_function画类似等高线的决策界面
```

```python
from sklearn.datasets import make_blobs
centers = [[1,2],[6,8]]
X,y = make_blobs(20,centers = centers)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure()
import numpy as np
for label,color,marker in zip(set(y),'by',['^','s']):
    index = np.where(y == label)
    plt.scatter(X[index,0],X[index,1],c = color,marker = marker)
plt.savefig('make_blobs.jpg')#在show之前保存
plt.show()
```

```python
from sklearn.svm import LinearSVC 
clf = LinearSVC()
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf,X,y,scoring = 'accuracy')
print(np.mean(scores))
```

```python
from sklearn.datasets import make_blobs
centers = [[1,2],[1,2]]
X,y = make_blobs(20,centers = centers,cluster_std = [1,4])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure()
import numpy as np
for label,color,marker in zip(set(y),'by',['^','s']):
    index = np.where(y == label)
    plt.scatter(X[index,0],X[index,1],c = color,marker = marker)
```

```python
from sklearn.svm import SVC 
clf = SVC()
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf,X,y,scoring = 'accuracy')
print(np.mean(scores))
```

```python
clf = SVC()
clf.fit(X,y)
X1_min,X1_max = X[:,0].min()-1,X[:,0].max()+1
X2_min,X2_max = X[:,1].min()-1,X[:,1].max()+1
xx,yy = np.meshgrid(np.arange(X1_min,X1_max,0.02),np.arange(X2_min,X2_max,0.02))
Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx,yy,Z,cmap = plt.cm.Paired)
for label,color,marker in zip(set(y),'by',['^','s']):
    index = np.where(y == label)
    plt.scatter(X[index,0],X[index,1],c = color,marker = marker)
plt.figure(2)
plt.contourf(xx,yy,Z,cmap = plt.cm.Paired)
for label,color,marker in zip(set(y),'by',['^','s']):
    index = np.where(y == label)
    plt.scatter(X[index,0],X[index,1],c = color,marker = marker)
```
