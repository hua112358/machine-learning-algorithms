```python
#导入数据
import numpy as np
np.random.seed(0)
X = np.r_[np.random.randn(20,2)-[2,2],np.random.randn(20,2)+[2,2]]
y = np.r_[[0]*20,[1]*20]
```

```python
#可视化
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure()
for label,color,marker in zip(set(y),'by',['o','s']):
    index = np.where(y==label)
    plt.scatter(X[index,0],X[index,1],color = color,marker = marker)
from sklearn import svm
clf = svm.SVC(kernel = 'linear')
clf.fit(X,y)
coef = clf.coef_[0]
intercept = clf.intercept_[0]
xmin,xmax = plt.xlim()
ymin = (-coef[0]*xmin-intercept)/coef[1]
ymax = (-coef[0]*xmax-intercept)/coef[1]
plt.plot([xmin,xmax],[ymin,ymax],'k-')
support_vector = clf.support_vectors_
xydown = support_vector[0]
xyup = support_vector[-1]
ydmin = (-coef[0]*xmin-(-coef[0]*xydown[0]-coef[1]*xydown[1]))/coef[1]
ydmax = (-coef[0]*xmax-(-coef[0]*xydown[0]-coef[1]*xydown[1]))/coef[1]
yumin = (-coef[0]*xmin-(-coef[0]*xyup[0]-coef[1]*xyup[1]))/coef[1]
yumax = (-coef[0]*xmax-(-coef[0]*xyup[0]-coef[1]*xyup[1]))/coef[1]
plt.plot([xmin,xmax],[ydmin,ydmax],'k--')
plt.plot([xmin,xmax],[yumin,yumax],'k--')
# coef = clf.coef_[0]
# intercept = clf.intercept_[0]
# support_vector = clf.support_vectors_
# points_x = np.linspace(-5,5)
# points_y = (-intercept-coef[0]*points_x)/coef[1]
# xydown = support_vector[0]
# xyup = support_vector[-1]
# intercept_down = (-coef[0]*xydown[0]-coef[1]*xydown[1])
# intercept_up = (-coef[0]*xyup[0]-coef[1]*xyup[1])
# points_yd = (-intercept_down-coef[0]*points_x)/coef[1]
# points_yu = (-intercept_up-coef[0]*points_x)/coef[1]
# plt.plot(points_x,points_y,'k-')
# plt.plot(points_x,points_yd,'k--')
# plt.plot(points_x,points_yu,'k--')
plt.axis('tight')
plt.show()
```
