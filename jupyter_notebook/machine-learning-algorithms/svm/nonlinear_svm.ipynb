```python
#导入数据
import numpy as np
np.random.seed(0)
X = np.random.randn(300,2)
y = np.logical_xor(X[:,0]>0,X[:,1]>0)
```

```python
%matplotlib inline
from sklearn import svm
import matplotlib.pyplot as plt
clf = svm.NuSVC()
clf.fit(X,y)
xx,yy = np.meshgrid(np.linspace(-3,3,500),np.linspace(-3,3,500))
Z = clf.decision_function(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(0)
plt.imshow(Z,extent = (xx.min(),xx.max(),yy.min(),yy.max()),
           aspect = 'auto',origin = 'lower',cmap = plt.cm.PuOr_r)
plt.contour(xx,yy,Z,levels = [0],linewidths = 2,linetypes = '--')
plt.scatter(X[:,0],X[:,1],s = 30,c = y,cmap = plt.cm.Paired)
plt.xlim(-3,3)
plt.ylim(-3,3)
# plt.axis([-3,3,-3,3])
plt.figure(1)
plt.contour(xx,yy,Z,cmap = plt.cm.Paired)#画等值线不填充颜色
plt.figure(2)
plt.contourf(xx,yy,Z,cmap = plt.cm.PuOr_r)#画等值线并填充颜色
plt.figure(0)
plt.savefig('nonlinear_plot1.jpg')
plt.figure(1)
plt.savefig('nonlinear_plot2.jpg')
plt.figure(2)
plt.savefig('nonlinear_plot3.jpg')
plt.show()
```
