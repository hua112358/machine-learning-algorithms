```python
#导入数据
from sklearn import datasets
dataset = datasets.load_iris()
X = dataset.data[:,:2]
y = dataset.target
```

```python
#数据预处理
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler()
X_transformed = transformer.fit_transform(X)
```

```python
#参数设置
from matplotlib.colors import ListedColormap
n_neighbors = 15
grid_color = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
points_color = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
h = 0.02
```

```python
#构建模型、可视化
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
for weights in ['uniform','distance']:
    clf = KNeighborsClassifier(n_neighbors = n_neighbors,weights = weights) 
    clf.fit(X,y)
    X1_min,X1_max = X[:,0].min() - 1,X[:,0].max() + 1
    X2_min,X2_max = X[:,1].min() - 1,X[:,1].max() + 1
    xx,yy = np.meshgrid(np.arange(X1_min,X1_max,h),np.arange(X2_min,X2_max,h))
    Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    Z = Z.reshape(xx.shape)
    fig = plt.figure()
    plt.pcolormesh(xx,yy,Z,cmap = grid_color)
    plt.scatter(X[:,0],X[:,1],c = y,cmap = points_color)
    plt.xlim(X1_min,X1_max)
    plt.ylim(X2_min,X2_max)
    plt.title('3-classes classification (k = %i, weights = %s)' % (n_neighbors,weights))
```
