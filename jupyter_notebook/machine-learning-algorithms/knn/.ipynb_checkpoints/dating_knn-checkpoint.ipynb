```python
#导入数据
import pandas as pd
dating_dataset = pd.read_csv('dating_dataset.csv',header = None,delimiter = '\t',names = ['flying_miles','gametime_percentage','icecream','class'])
```

```python
#两种方式创建数据集
# import numpy as np
# X = np.array(dating_dataset.iloc[:,:-1])
# y = np.array(dating_dataset.iloc[:,-1])
X = dating_dataset[['flying_miles','gametime_percentage','icecream']].values
y = dating_dataset['class'].values
```

```python
#数据可视化
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure()
plt.subplot(111)
plt.scatter(X[:,1],X[:,2],15.0*y,15.0*y)
```

```python
#数据预处理
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler()
X_transformed = transformer.fit_transform(X)
```

```python
#构建、评估分类器
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import cross_val_score
clf = KNeighborsClassifier(n_neighbors = 5)
scores = cross_val_score(clf,X_transformed,y,scoring = 'accuracy')
print('the average score is {0:.1f}%'.format(np.mean(scores)*100))
```

```python
#用分类器进行预测
clf.fit(X_transformed,y)
y_predicted = clf.predict(X_transformed)
print(np.mean(y == y_predicted))
print(clf.score(X_transformed,y))
```

```python
from sklearn.model_selection import GridSearchCV
clf = KNeighborsClassifier()
Cs = np.logspace(-6, -1, 10)
clf = GridSearchCV(estimator = clf,param_grid = dict(C = Cs))
clf.fit(X,y)
```
