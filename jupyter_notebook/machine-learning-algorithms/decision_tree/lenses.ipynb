```python
#导入数据
with open('lenses.txt','r') as f:
    lenses_dataset = f.readlines()
X = []
y = []
for instance in lenses_dataset:
    instance = instance.strip().split('\t')
    X.append(instance[:-1])
    y.append(instance[-1])
feature_names = ['age','prescript','astigmatic','tear_rate']
```

```python
#数据预处理
from sklearn.feature_extraction import DictVectorizer
X_dict_list = []
for instance in X:
    instance_dict = {}
    for i in range(len(instance)):
        instance_dict[feature_names[i]] = instance[i]
    X_dict_list.append(instance_dict)
dict_vectorizer_transformer = DictVectorizer()
X_transformed = dict_vectorizer_transformer.fit_transform(X_dict_list).toarray()
from sklearn.preprocessing import LabelEncoder
label_encoder_transformer = LabelEncoder()
y_transformed = label_encoder_transformer.fit_transform(y)
feature_names_transformed = dict_vectorizer_transformer.get_feature_names()
class_names_transformed = label_encoder_transformer.classes_
```

```python
#构建、评估模型
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
decision_tree_clf = DecisionTreeClassifier()
scores = cross_val_score(decision_tree_clf,X_transformed,y,scoring = 'accuracy')
print('the average score is {0:.1f}%'.format(np.mean(scores)*100))
```

```python
#结果可视化
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
decision_tree_clf = DecisionTreeClassifier()
decision_tree_clf.fit(X_transformed,y_transformed)
dot_data = tree.export_graphviz(decision_tree_clf,out_file = None,feature_names = feature_names_transformed,
                               class_names = class_names_transformed,rounded = True,filled = True)
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
from IPython.display import Image
Image(graph.create_gif())
```

```python
#预测
decision_tree_clf.predict(np.array([1,1,1,1,1,1,1,1,1]).reshape(1,-1))
```
