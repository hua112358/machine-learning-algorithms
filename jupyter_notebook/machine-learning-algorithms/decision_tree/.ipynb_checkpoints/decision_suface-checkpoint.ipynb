```python
#导入数据
from sklearn import datasets
iris_dataset = datasets.load_iris()
X = iris_dataset.data
y = iris_dataset.target
```

```python
#画出决策面
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
grid_color = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
points_color = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
n_classes = 3
plot_colors = 'bry'
plot_step = 0.02
for pairidx,pair in enumerate([[0,1],[0,2],[0,3],[1,2],[1,3],[2,3]]):
    X_train = X[:,pair]
    y_train = y
    from sklearn.tree import DecisionTreeClassifier
    decision_tree_clf = DecisionTreeClassifier()
    decision_tree_clf.fit(X_train,y_train)
    X1_min,X1_max = X_train[:,0].min() - 1,X_train[:,0].max() + 1
    X2_min,X2_max = X_train[:,1].min() - 1,X_train[:,1].max() + 1
    import numpy as np
    xx,yy = np.meshgrid(np.arange(X1_min,X1_max,plot_step),np.arange(X2_min,X2_max,plot_step))
    Z = decision_tree_clf.predict(np.c_[xx.ravel(),yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.subplot(2,3,pairidx + 1)
    plt.pcolormesh(xx,yy,Z,cmap = plt.cm.Paired)
    #cs = plt.contourf(xx, yy, Z, cmap = plt.cm.Paired)
    plt.xlabel(iris_dataset.feature_names[pair[0]])
    plt.ylabel(iris_dataset.feature_names[pair[1]])
    plt.axis('tight')
#     plt.scatter(X_train[:,0],X_train[:,1],c = y,cmap = points_color)
#     plt.xlim(X1_min,X1_max)
#     plt.ylim(X2_min,X2_max)
    for i,color in zip(range(n_classes),plot_colors):
        idx = np.where(y==i)
        plt.scatter(X_train[idx,0],X_train[idx,1],c = color,label = iris_dataset.target_names[i],cmap = 
                   points_color)
    plt.axis('tight')
plt.suptitle("Decision surface of a decision tree using paired features")
plt.legend()
```
