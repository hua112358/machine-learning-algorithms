```python
import re
pattern = re.compile('<tr.*?homesxname=(.*?)awaysxname=(.*?) class=.*?target="_blank">(.*?)</a>.*?<a class="score".*?>(.*?)</a>.*?value="3".*?>(.*?)<.*?value="1".*?>(.*?)<.*?value="0".*?>(.*?)<',re.S)
items = re.findall(pattern,content)
results = []
for item in items:
    if item[2] == '日职乙':
        print(item)
        score = item[3].split(':')
        if int(score[0])<int(score[1]):
            results.append([float(item[4]),float(item[5]),float(item[6]),0])
        elif int(score[0])==int(score[1]):
            results.append([float(item[4]),float(item[5]),float(item[6]),1])
        else:
            results.append([float(item[4]),float(item[5]),float(item[6]),3])
```

```python
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
content = response.read().decode('gb2312')
results = []
for date in range(1,31):
    url = ('http://trade.500.com/jczq/?date=2016-06-%i&playtype=both' % date)
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    content = response.read().decode('gb2312')
    import re
    pattern = re.compile('<tr.*?homesxname=(.*?)awaysxname=(.*?) class=.*?target="_blank">(.*?)</a>.*?<a class="score".*?>(.*?)</a>.*?value="3".*?>(.*?)<.*?value="1".*?>(.*?)<.*?value="0".*?>(.*?)<',re.S)
    items = re.findall(pattern,content)
    for item in items:
        if (item[2] == '日职乙') or (item[2] == '日职'):
            print(item)
            score = item[3].split(':')
            if int(score[0])<int(score[1]):
                results.append([float(item[4]),float(item[5]),float(item[6]),0,int(score[0])-int(score[1])])
            elif int(score[0])==int(score[1]):
                results.append([float(item[4]),float(item[5]),float(item[6]),1,int(score[0])-int(score[1])])
            else:
                results.append([float(item[4]),float(item[5]),float(item[6]),3,int(score[0])-int(score[1])])
```
