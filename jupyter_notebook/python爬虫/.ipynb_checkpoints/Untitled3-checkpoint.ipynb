```python
import urllib.request
results = []
for month in range(1,13):
    for date in range(1,31):
        url = ('http://trade.500.com/jczq/?date=2016-%i-%i&playtype=both' % (month,date))
        request = urllib.request.Request(url)
        response = urllib.request.urlopen(request)
        content = response.read().decode('gb2312')
        import re
        pattern = re.compile('<tr.*?pname=.*?pdate=(.*?) lg=.*?homesxname=(.*?)awaysxname=(.*?) class=.*?target="_blank">(.*?)</a>.*?<a class="score".*?>(.*?)</a>.*?value="3".*?>(.*?)<.*?value="1".*?>(.*?)<.*?value="0".*?>(.*?)<',re.S)
        items = re.findall(pattern,content)
        for item in items:
            if (item[3] == '日职乙') or (item[3] == '日职'):
                print(item)
                score = item[4].split(':')
                if int(score[0])<int(score[1]):
                    results.append([float(item[5]),float(item[6]),float(item[7]),0,int(score[0])-int(score[1])])
                elif int(score[0])==int(score[1]):
                    results.append([float(item[5]),float(item[6]),float(item[7]),1,int(score[0])-int(score[1])])
                else:
                    results.append([float(item[5]),float(item[6]),float(item[7]),3,int(score[0])-int(score[1])])
```

```python
results
```

```python
import numpy as np
results_array = np.array(results)
```

```python
results_array.shape
```

```python
#导入数据
import numpy as np
dataset = results_array
X = dataset[:,0:3]
y = dataset[:,3]
```

```python
#预处理
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler()
X_transformed = transformer.fit_transform(X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
clf_knn = KNeighborsClassifier()
scores = cross_val_score(clf_knn,X,y,scoring = 'accuracy')
print('the average score is {0:.1f}%'.format(np.mean(scores)*100))
scores_transformed = cross_val_score(clf_knn,X_transformed,y,scoring ='accuracy')
print('the average score of transformed is {0:.1f}%'.format(np.mean(scores_transformed)*100))
```

```python
from sklearn.linear_model import LogisticRegression
clf_logistic = LogisticRegression()
scores = cross_val_score(clf_logistic,X,y,scoring = 'accuracy')
print('the average score is {0:.1f}%'.format(np.mean(scores)*100))
scores_transformed = cross_val_score(clf_logistic,X_transformed,y,scoring ='accuracy')
print('the average score of transformed is {0:.1f}%'.format(np.mean(scores_transformed)*100))
```

```python
from sklearn.naive_bayes import GaussianNB
clf_bayes = GaussianNB()
scores = cross_val_score(clf_bayes,X,y,scoring = 'accuracy')
print('the average score is {0:.1f}%'.format(np.mean(scores)*100))
scores_transformed = cross_val_score(clf_bayes,X_transformed,y,scoring ='accuracy')
print('the average score of transformed is {0:.1f}%'.format(np.mean(scores_transformed)*100))
```

```python
from sklearn.svm import SVC
clf_svm = SVC()
scores = cross_val_score(clf_svm,X,y,scoring = 'accuracy')
print('the average score is {0:.1f}%'.format(np.mean(scores)*100))
scores_transformed = cross_val_score(clf_svm,X_transformed,y,scoring ='accuracy')
print('the average score of transformed is {0:.1f}%'.format(np.mean(scores_transformed)*100))
```
